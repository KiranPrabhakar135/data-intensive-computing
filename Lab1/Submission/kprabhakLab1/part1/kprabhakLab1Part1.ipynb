```r
sales1<-c(12,14,16,29,30,45,19,20,16, 19, 34, 20) # creates a list with the given data
sales2<-rpois(12,34)
par(bg="cornsilk")
# plots the graph with sales data and the plot will have y-axis limit from 0-100
plot(sales1, col = "blue", type = "o", ylim = c(0,100), xlab = "Month", ylab = "Sales") 
# The rpois function will create random numbers between 12 and 34 using poission distribution. So this is the reason
# the second plot will appear differently when the code is run multiple times.
sales2<-rpois(12,34)
par(bg="cornsilk")
# plots the graph and connects the points with lines
lines(sales2, type = "o", lty = 2, pch=22, col="red")
grid(nx=NA, ny=NULL)
legend("topright", inset=.05, c("Sales1","Sales2"), fill=c("blue","red"), horiz=TRUE)
```

```r
# opens a prompt to choose the file and reads the table in the text file
sales <- read.table('salesData.txt',header = T)
sales
# deside 'T'(True) will plot the graphs adjacent to ecah other.
barplot(as.matrix(sales), main="Sales Data", ylab= "Total",beside=T, col=rainbow(5))
# From the plot it is evident that the value of (12,80) and (23,89) are (min,max) values for sales1 and sales2 data
```

```r
# The blow plot shows the median values fot the two sales details, the lower whisker and the upper whisker shows the lowest
# and highest values of sales data. The lower hinge indicates the value of 1st quartile and the upper hinge mentiones the value
# of second quartile.
sales
# Gives a matrix, where each column contains the extreme of the lower whisker, the lower hinge, the median, 
# the upper hinge and the extreme of the upper whisker for one group/plot.
fn<-boxplot(sales,col=c("orange","green"))$stats
# fn is a matrix with 5 rows and two columns [3,2] and [3,1] are the medians which is 34 for sales1 and 45 for sales2
# The first and third quartile values are (26, 55.5) for the sales1 and (35 and 58) for sales2.
# The 1.45 and 0.45 values are to adjust the postion of text and the 'cex' value will adjust the font of the text
text(1.45, fn[3,2], paste("Median =", fn[3,2]), adj=0, cex=.7)
text(0.45, fn[3,1], paste("Median =", fn[3,1]), adj=0, cex=.7)
grid(nx=NA, ny=NULL)
```

```r
fb1<-read.csv('FB.csv')
appl1<-read.csv('AAPL.csv')
par(bg="cornsilk")
# displays the stock details of Adj.Close of appl and fb
plot(appl1$Adj.Close, col="blue", type="o", ylim=c(150,200), xlab="Days", ylab="Price" )
lines(fb1$Adj.Close, type="o", pch=22, lty=2, col="red")
legend("topright", inset=.05, c("Apple","Facebook"), fill=c("blue","red"), horiz=TRUE)
# The below  lines of code draws the histograms for the stock prices of apple and facebook
hist(appl1$Adj.Close, col=rainbow(8), xlab="Adjusted Close", main='Histogram of Adjusted Close of Apple')
# The histogram indicates the apple stock price is around (170 - 178) most of the cases and (174-176) is most frequest price
hist(fb1$Adj.Close, col=rainbow(8), xlab="Adjusted Close", main='Histogram of Adjusted Close of Facebook')
# The histogram indicates the facebook stock price is around (175 - 180) most of the time.
```
